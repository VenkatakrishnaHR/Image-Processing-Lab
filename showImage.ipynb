```python
import cv2
from google.colab.patches import cv2_imshow
```

Show Image

```python
img = cv2.imread('image.jpeg')
print(type(img))

# cv2.imshow(img)
cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Show image in gray scale mode

```python
#Here, 0 indicates gray scale level
img = cv2.imread("image.jpeg", 0)

# cv2.imshow(img)
cv2_imshow(img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
