# Write a python program to load all the images stored in a directory

```python
import cv2
import os
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
directory = "imageDirectory"

images = []

for file in os.listdir( directory ):
  images.append( cv2.imread( directory + os.sep + file ) )

for image in images:
  cv2_imshow(image)
```

# Resize all images to same size

```python
height, width = images[0].shape[:2]
for i in range( 0, len(images) ):
  images[i] = cv2.resize(images[i],(width, height), interpolation = cv2.INTER_CUBIC)
for image in images:
  cv2_imshow(image)
```

```python
meanMatrix = np.mean( images, 0 )
meanMatrix.shape
cv2_imshow( meanMatrix )
```

```python
summedImage = np.sum(images[-3:-1], 0)
summedImage.shape
cv2_imshow( summedImage )
```

```python
img1 = images[-3]
img2 = images[-2]
dst = cv2.addWeighted(img1,0.3,img2,0.7,0)
cv2_imshow( dst )
```
