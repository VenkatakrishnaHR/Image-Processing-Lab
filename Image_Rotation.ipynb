#Develope program to perform linear rotation on an image

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('image.jpeg')
cv2_imshow( image )
```

```python
image.shape
```

```python
height, width = image.shape[:2]
  
# get the center coordinates of the image to create the 2D rotation matrix
# center = (width/2, height/2)
center = (max(width/2, height/2), max(width/2, height/2))
  
# using cv2.getRotationMatrix2D()  to get the rotation matrix
rotate_matrix = cv2.getRotationMatrix2D(center=center, angle=90, scale=1)
  
# rotate the image using cv2.warpAffine 90 degree anticlockwise
rotated_image = cv2.warpAffine( src=image, M=rotate_matrix, dsize=(height, width))
cv2_imshow( rotated_image )
```
