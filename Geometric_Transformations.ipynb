```python
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

```python
img = cv.imread('image.jpeg')
plt.imshow( img )
```

1. Scalling (Resizing)

```python
#Method 1: Size spcified in scalling factors
res = cv.resize(img,None,fx=2, fy=2, interpolation = cv.INTER_LINEAR) # or INTER_AREA or shrinking

plt.subplot( 1, 2, 1 )
plt.title( "Before Scalling" )
plt.imshow( img )

plt.subplot( 1, 2, 2 )
plt.title( "After Scalling" )
plt.imshow( res )
```

```python
#Method 1: Size spcified manually in pixels
height, width = img.shape[:2]
res = cv.resize(img,(2*width, 2*height), interpolation = cv.INTER_CUBIC)


plt.subplot( 1, 2, 1 )
plt.title( "Before Scalling" )
plt.imshow( img )

plt.subplot( 1, 2, 2 )
plt.title( "After Scalling" )
plt.imshow( res )
```

2. Translation

```python
rows,cols = img.shape[:2]
M = np.float32([[1,0,rows/2],[0,1, rows/3 ]])
dst = cv.warpAffine(img,M,(cols,rows))

plt.subplot( 1, 2, 1 )
plt.title( "Before translation" )
plt.imshow( img )

plt.subplot( 1, 2, 2 )
plt.title( "After translation" )
plt.imshow( dst )
```

3. Rotation

```python
rows,cols = img.shape[:2]
# cols-1 and rows-1 are the coordinate limits.
M = cv.getRotationMatrix2D(((cols-1)/3.0,(rows-1)/2.0),45,1.5)
dst = cv.warpAffine(img,M,(cols,rows))


plt.subplot( 121 )
plt.title( "Before Rotation" )
plt.imshow( img )

plt.subplot( 122 )
plt.title( "After Rotation" )
plt.imshow( dst )
```
